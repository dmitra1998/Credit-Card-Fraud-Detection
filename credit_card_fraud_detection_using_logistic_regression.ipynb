```python
#Importing dependencies
import numpy as np
import pandas as pd  #numpy and pandas are library
from sklearn.model_selection import train_test_split #it will split data in test and train
from sklearn.linear_model import LogisticRegression #model to be used in our project
from sklearn.metrics import accuracy_score ##will help in checking in accuracy
```

```python
# loading the dataset to a panda dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first five rows
credit_card_data.head()
```

```python
#last five rows
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the number of missing value in each column
credit_card_data.isnull().sum()
```

```python
credit_card_data.shape
```

```python
#Replacing missing value with null string method called imputation
credit_card_data = credit_card_data.fillna('')
```

```python
#distribution of legit transaction & fraudulent transactions
credit_card_data['Class'].value_counts()
#This dataset is unbalanced as valid transaction(0.0) is much more the fraudulant transaction(1.0)
```

```python
#seperating the data for analysis
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
#here it is percentile not percentage 25, 50 and 75
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

Under-Sampling will be used to fix unbalanced dataset

Build a sample dataset containing similar distribution of normal transaction and fraudulent transaction

Number of fraudulent transaction --> 392

```python
legit_sample = legit.sample(n=392) #it will choose 392 random data from legit variable
```

Concatenating two dataframe

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0) #axis=0 means value will be added row wise and axis=1 means value will be added column wise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
#total values present in each class
new_dataset['Class'].value_counts()
```

```python
#compare the values for both transaction for new dataset
new_dataset.groupby('Class').mean()
```

Splitting data into Features & Target

```python
#Feature are all columns
#Target are rows (here 0 and 1)
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X
```

```python
Y
```

Splitting data into training and testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression will be used for binary represented data

```python
model = LogisticRegression()
```

```python
#training data with LogisticRegression with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Testing data : ', test_data_accuracy)
```

If accuracy score of training and test data are very different then our model is either overfitted or underfitted. 

Suppose if training data is having accuracy of 94% and test data have accuracy of 50% then the model is overfitted and vice-versa for underfitting
